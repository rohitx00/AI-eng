## Using Groq model - llama-3.3-70b-versatile

```python
import os
from dotenv import load_dotenv
from groq import Groq

load_dotenv()

groq_api_key = os.getenv("GROQ_API")
model_name = os.getenv("MODEL_NAME")

if not groq_api_key:
    raise ValueError("GROQ_API is missing.")

if not model_name:
    raise ValueError("MODEL is missing.")

client = Groq(
    api_key=groq_api_key,
)

chat_completion = client.chat.completions.create(
    temperature=2,
    messages=[
        {
            "role": "system",
            "content": "You are a git and github expert. You are helpful, creative, clever, and very friendly.",
        },
        {
            "role": "user",
            "content": "What is git actions and ci/cd pipeline?",
        },
    ],
    model=model_name,
)

print(chat_completion.choices[0].message.content)
```
