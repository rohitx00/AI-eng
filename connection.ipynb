## Connecting with Groq

```python
import os
from pathlib import Path
from dotenv import load_dotenv
from groq import Groq

load_dotenv()

groq_api_key = os.getenv("GROQ_API")

if not groq_api_key:
    raise ValueError("GROQ_API is missing.")

client = Groq(
    api_key=groq_api_key,
)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": "write a python function that takes a list of integers and returns the sum of all even numbers in the list",
        },
        {
            "role": "user",
            "content": "Here is a Python function that takes a list of integers and returns the sum of all even numbers in the list:",
        },
    ],
    model="llama-3.3-70b-versatile",
)

print(chat_completion.choices[0].message.content)
```
